```python
def integer_length_vectors():
    for i in range(1, 100):
        for j in range(1, i+1):
            for k in range(1, j+1):
                v = (i, j, k)
                if length(v).is_integer():
                    yield v, length(v)
                    
for v, l in integer_length_vectors():
    print(v, l)
```

```python
def render(faces, light_source, colormap = matplotlib.cm.get_cmap('Blues')):
    polygons = []
    unit_light_source = unit(light_source)
    for face in faces:
        unit_normal = unit(normal(face))
        if (unit_normal[2] > 0): # draw only feces turned towards positive Z
            projected_face = project_vectors_to_xy_plane(*face)
            fill = colormap(1 - dot(unit_normal, unit_light_source))
            polygons.append(d2.Polygon2D(*projected_face, fill=fill, color=None))
            
    print(len(polygons))
    d2.draw2d(*polygons, axes=False, origin=False, grid=None)
            
render(octahedron, (1, 2, 3), colormap=matplotlib.cm.get_cmap('Reds'))
```

```python
def split(face):
    midpoints = [unit(add(face[i], face[(i+1)%len(face)])) for i in range(len(face))]
    triangles = [[face[i], midpoints[i], midpoints[(i-1)%len(face)]] for i in range(len(face))]
    return [midpoints] + triangles

def recursive_split(faces=octahedron, depth=1):
    if depth == 0:
        return faces
    else:
        return recursive_split([new_face for face in faces for new_face in split(face)], depth-1)
    
render(recursive_split(octahedron, 1), (1, 2, 3), colormap=matplotlib.cm.get_cmap('Blues'))
render(recursive_split(octahedron, 2), (1, 2, 3), colormap=matplotlib.cm.get_cmap('Greens'))
render(recursive_split(octahedron, 5), (1, 2, 3), colormap=matplotlib.cm.get_cmap('Purples'))
```

```python
def segments(*faces):
    return list(set([d3.Segment3D(face[(i+1)%len(face)], face[i])
                     for face in faces for i in range(len(face))]))


d3.draw3d(*segments(*octahedron))
```
